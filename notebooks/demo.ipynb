```python
import sys
sys.path.append("..")
from spark_session import get_spark

spark = get_spark()

# Delta Lake
df = spark.createDataFrame([(1, "Alice"), (2, "Bob")], ["id", "name"])
df.write.format("delta").mode("overwrite").save("warehouse/delta/people")

spark.read.format("delta").load("warehouse/delta/people").show()

# Iceberg
spark.sql("CREATE TABLE IF NOT EXISTS local.db.people (id INT, name STRING) USING iceberg")
df.writeTo("local.db.people").append()
spark.table("local.db.people").show()
```
